```python
country = "NO"

query_url = f"https://api.chess.com/pub/country/{country}/players"

print(f"Fetching player list for country {country}")

username_NO = requests.get(query_url)
username_NO_json = username_NO.json()
username_NO_json
```

```python
usernames_NO = []

for username in username_NO_json["players"]:
    usernames_NO.append(username)

player_NO_dict = {"username": usernames_NO}

chessplayer_NO_df = pd.DataFrame(player_NO_dict)
chessplayer_NO_df.to_csv(output_data_players_NO, index=True)
chessplayer_NO_df
```
